```python
import hashlib
country = "CAN"
df.loc[df['NOC'] == country]['Name'].apply(lambda x: hashlib.sha256(x.encode('utf-8')).hexdigest())

df_country = df.loc[df["NOC"] == country]
```

```python
df_sports = df_country.loc[(df_country["Medal"] == "Gold") | (df_country["Medal"] == "Silver") | (df_country["Medal"] == "Bronze")].drop_duplicates(
    subset=["Event", "Year"])


fig = px.pie(df_sports, values=df_sports["Medal"].value_counts(
), names=df_sports["Medal"].value_counts().index, title=f"Medals won by {country}")
fig.update_traces(
    hovertemplate="Medal: %{label}: <br>Number of medals: %{value}")
fig.show()
```

```python
df_sports = df_country.drop_duplicates(subset=["Event", "Year"])

df_sports_medals = dict(df_sports.groupby("Sport")["Medal"].count())

a = sorted(df_sports_medals.items(), key=lambda x: x[1], reverse=True)

top_ten_sports = [a[i][0] for i in range(10) ]

top_ten_sports_medals = [a[i][1] for i in range(10)]

fig = px.bar(x=top_ten_sports, y=top_ten_sports_medals)


fig.show()
```

### Sportstatistik

medaljfördelning mellan länder i sporterna

```python
def medal_distribution_countries(sport: str)->None:
    df_sports = df.loc[df["Sport"] == sport].groupby("NOC")["Medal"].count()

    df_sports = df_sports.sort_values(ascending=False)


    fig = px.bar(df_sports.head()) 
    fig.show()
    

```
